# Pytorch Code-V1

```python
import torch
import torch.optim as optim
```

```python
x1 = torch.FloatTensor([[20], [19], [18.5], [20], [14.5]])
x2 = torch.FloatTensor([[15], [18], [19], [20], [13]])
x3 = torch.FloatTensor([[17], [18.5], [18], [20], [15.5]])
y = torch.FloatTensor([[87], [92.5], [91], [100], [72]])

w1 = torch.zeros(1, requires_grad=True)
w2 = torch.zeros(1, requires_grad=True)
w3 = torch.zeros(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)

optimizer = optim.SGD([w1, w2, w3, b], lr=1e-3)

for epochs in range(1001):
  hypothesis = w1*x1 + w2*x2 + w3*x3 + b
  cost = torch.mean((hypothesis - y)**2)

  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  if epochs % 100 == 0:
    print("Epoch:  {:3d} / {} W1: {:.5f} W2: {:.5f} W3: {:.5f} b: {:.5f} Hypothesis: {} \
     Cost: {:.5f}".format(epochs, 1000, w1.item(), w2.item(), w3.item(), \
                          b.item(), hypothesis.detach(), cost.item()))
```

# PyTorch Code-V2

```python
x = torch.FloatTensor([[20, 15, 17],
                       [19, 18, 18.5],
                       [18.5, 19, 18],
                       [20, 20, 20],
                       [14.5, 13, 15.5]]) #Shape of X: 5X3
y = torch.FloatTensor([[87],
                       [92.5],
                       [91],
                       [100],
                       [72]]) #Shape of y: 5X1

w = torch.zeros((3, 1), requires_grad = True)
b = torch.zeros(1, requires_grad = True)

optimizer = optim.SGD([w, b], lr=1e-3)

for epochs in range(1001):
  hypothesis = x.matmul(w) + b #matmul method is for matrix multiplication
  cost = torch.mean((hypothesis - y)**2)

  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  if epochs % 100 == 0:
    print("Epoch:  {:3d} / {} W1: {:.5f} W2: {:.5f} W3: {:.5f} b: {:.5f} Hypothesis: {} \
     Cost: {:.5f}".format(epochs, 1000, w1.item(), w2.item(), w3.item(), \
                          b.item(), hypothesis.detach(), cost.item()))
```

# TensorFlow Code-V1

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

x1 = [20, 19, 18.5, 20, 14.5]
x2 = [15, 18, 19, 20, 13]
x3 = [17, 18.5, 18, 20, 15.5]
y = [87, 92.5, 91, 100, 72]

w1= tf.Variable(tf.random_normal([1], name="w1"))
w2= tf.Variable(tf.random_normal([1], name="w2"))
w3= tf.Variable(tf.random_normal([1], name="w3"))
b= tf.Variable(tf.random_normal([1], name="bias"))

hypothesis = w1*x1 + w2*x2 + w3*x3 + b
cost = tf.reduce_mean(tf.square(hypothesis - y))

# optimizer = tf.train.GradientDescentOptimizer(learning_rate = 1e-3)
# train = optimizer.minimize(cost)
train = tf.train.GradientDescentOptimizer(learning_rate = 1e-3).minimize(cost)

#Run a graph
# with tf.Session() as sess:
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for epoch in range(1001):
  _, w1_val, w2_val, w3_val, b_val, cost_val, hy_val = sess.run([train, w1, w2, w3, b, cost, hypothesis])

  if epoch % 100 == 0:
    print(epoch, w1_val, w2_val, w3_val, b_val, cost_val, hy_val)
```

# TensorFlow Code-V2

```python
x1 = tf.placeholder(tf.float32, shape=None) #None means let tensorflow decide a dataset shape automatically
x2 = tf.placeholder(tf.float32, shape=None)
x3 = tf.placeholder(tf.float32, shape=None)
y = tf.placeholder(tf.float32, shape=None)

w1 = tf.Variable(tf.random_normal([1], name="w1"))
w2 = tf.Variable(tf.random_normal([1], name="w2"))
w3 = tf.Variable(tf.random_normal([1], name="w3"))
b = tf.Variable(tf.random_normal([1], name="bias"))

hypothesis = w1*x1 + w2*x2 + w3*x3 + b
cost = tf.reduce_mean(tf.square(hypothesis - y))

train = tf.train.GradientDescentOptimizer(learning_rate = 1e-3).minimize(cost)

# with tf.Session() as sess: #keyword with is optional
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for epoch in range(1001):
  _, w1_val, w2_val, w3_val, b_val, cost_val, hy_val = \
  sess.run([train, w1, w2, w3, b, cost, hypothesis],\
           feed_dict={x1: [20, 19, 18.5, 20, 14.5],
            x2: [15, 18, 19, 20, 13],
            x3: [17, 18.5, 18, 20, 15.5],
            y: [87, 92.5, 91, 100, 72]})
  if epoch % 100 == 0:
    print(epoch, w1_val, w2_val, w3_val, b_val, cost_val, hy_val)
```

# TensorFlow with Pandas and Numpy

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame([{'x1': 20, 'x2': 15, 'x3': 17, 'y': 87},
                   {'x1': 19, 'x2': 18, 'x3': 18.5, 'y': 92.5},
                   {'x1': 18.5, 'x2': 19, 'x3': 18, 'y': 91},
                   {'x1': 20, 'x2': 20, 'x3': 20, 'y': 100},
                   {'x1': 14.5, 'x2': 13, 'x3': 15.5, 'y': 72}])
df
```

```python
x_data = df.iloc[:, 0:-1].to_numpy() #Convert into numpy format
y_data = df.iloc[:, [-1]].to_numpy()

x_data
```

```python
x = tf.placeholder(tf.float32, shape=[None, None])
y = tf.placeholder(tf.float32, shape=[None, None])

w = tf.Variable(tf.random_normal([3, 1]))
b = tf.Variable(tf.random_normal([1]))

hypothesis = tf.matmul(x, w) + b

cost = tf.reduce_mean(tf.square(hypothesis - y))

train = tf.train.GradientDescentOptimizer(learning_rate = 1e-3).minimize(cost)

# with tf.Session() as sess: #keyword with is optional
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for epoch in range(1001):
  _, w_val, b_val, cost_val, hy_val = \
  sess.run([train, w, b, cost, hypothesis],\
           feed_dict={x: x_data, y: y_data})
  if epoch % 100 == 0:
    print(epoch, w_val, b_val, cost_val, hy_val)
```
