```python
import pandas as pd
import numpy as np
from sklearn import model_selection as ms
from sklearn.model_selection import train_test_split
from pandas.api.types import CategoricalDtype
```

```python
df = pd.read_csv('/content/drive/MyDrive/CSCE464/datasets/CICIDS.csv')
df
```

```python
target = df['Label']
x_train, x_test = train_test_split(df, test_size=0.2, shuffle=True, stratify=target, random_state=777)
x_train, x_val = train_test_split(x_train, test_size=0.1, shuffle=True, random_state=777)

x_train = x_train.reset_index(drop=True)
x_test = x_test.reset_index(drop=True)
x_val = x_val.reset_index(drop=True)
```

```python
class dataset:
  x_train = x_train
  x_train_category = x_train['Label'].copy()

  x_val = x_val
  x_val_category = x_val['Label'].copy()

  x_test = x_test
  x_test_category = x_test['Label'].copy()
```

```python
category_variables = ['Label']

for cv in category_variables:
  dataset.x_train[cv] = dataset.x_train[cv].astype('category')
  dataset.x_val[cv] = dataset.x_val[cv].astype('category')
  dataset.x_test[cv] = dataset.x_test[cv].astype('category')

  dataset.x_train[cv] = dataset.x_train[cv].astype(CategoricalDtype(categories= dataset.x_train[cv].cat.categories))
  dataset.x_val[cv] = dataset.x_val[cv].astype(CategoricalDtype(categories= dataset.x_val[cv].cat.categories))
  dataset.x_test[cv] = dataset.x_test[cv].astype(CategoricalDtype(categories= dataset.x_test[cv].cat.categories))
```

```python
attack_types = {
    'DDoS' : 'DDos',
    'BENIGN' : 'Benign'
}

is_attack = {
    'Benign' : 'Normal',
    'DDos' : 'Attack'
}
```

```python
dataset.x_train['type'] = dataset.x_train.Label.map(lambda x: attack_types[x])
dataset.x_train['is'] = dataset.x_train.type.map(lambda x: is_attack[x])
dataset.x_train_category['type'] = dataset.x_train_category.map(lambda x: attack_types[x])
dataset.x_train_category['is'] = dataset.x_train_category.type.map(lambda x: is_attack[x])

dataset.x_val['type'] = dataset.x_val.Label.map(lambda x: attack_types[x])
dataset.x_val['is'] = dataset.x_val.type.map(lambda x: is_attack[x])
dataset.x_val_category['type'] = dataset.x_val_category.map(lambda x: attack_types[x])
dataset.x_val_category['is'] = dataset.x_val_category.type.map(lambda x: is_attack[x])

dataset.x_test['type'] = dataset.x_test.Label.map(lambda x: attack_types[x])
dataset.x_test['is'] = dataset.x_test.type.map(lambda x: is_attack[x])
dataset.x_test_category['type'] = dataset.x_test_category.map(lambda x: attack_types[x])
dataset.x_test_category['is'] = dataset.x_test_category.type.map(lambda x: is_attack[x])
```

```python
dummy_variables_2labels = [*category_variables, 'is']
dummy_variables_5labels = [*category_variables, 'type']
attack_codes_2labels = {'Normal': 0, 'Attack': 1}
```

```python
class preprocessing:
  x_train_2labels = pd.get_dummies(dataset.x_train, columns=dummy_variables_2labels, prefix=dummy_variables_2labels)
  x_val_2labels = pd.get_dummies(dataset.x_val, columns=dummy_variables_2labels, prefix=dummy_variables_2labels)
  x_test_2labels = pd.get_dummies(dataset.x_test, columns=dummy_variables_2labels, prefix=dummy_variables_2labels)

  x_train_2labels_y = dataset.x_train['is'].copy()
  x_val_2labels_y = dataset.x_val['is'].copy()
  x_test_2labels_y = dataset.x_test['is'].copy()

  x_train_2labels.drop(['type', 'Label_BENIGN', 'Label_DDoS'], axis=1, inplace=True)
  x_val_2labels.drop(['type', 'Label_BENIGN', 'Label_DDoS'], axis=1, inplace=True)
  x_test_2labels.drop(['type', 'Label_BENIGN', 'Label_DDoS'], axis=1, inplace=True)

  x_train_2labels_y = x_train_2labels_y.map(lambda x: attack_codes_2labels[x])
  x_val_2labels_y = x_val_2labels_y.map(lambda x: attack_codes_2labels[x])
  x_test_2labels_y = x_test_2labels_y.map(lambda x: attack_codes_2labels[x])
```
