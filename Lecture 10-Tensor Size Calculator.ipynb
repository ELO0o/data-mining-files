#Exercise 1: Slide #7


*   Input size = 227x227
*   Output size = ?
* Kernel size = 11x11
* Stride = 4



```python
import torch
import torch.nn as nn
```

```python
input = torch.Tensor(1, 1, 227, 227) #batch_size: 1, input_channel: 1, input_size: 227x227
input.shape

conv = nn.Conv2d(1, 1, (11, 11), stride=4, padding = 0) #input channel: 1, output channel: 1, kernel size: 11x11
conv

output = conv(input)
output.shape #55x55
```

#Exercise #5, slide #13: MNIST CNN

```python
input = torch.Tensor(1, 1, 28, 28)
conv1 = nn.Conv2d(1, 32, 3, stride=1, padding=1)
pool = nn.MaxPool2d(2, 2) #kernel size:2, size: 2

conv2 = nn.Conv2d(32, 64, 3, stride=1, padding=1)
pool = nn.MaxPool2d(2, 2) #kernel size:2, size: 2

out1 = conv1(input)
out1 = pool(out1)
out2 = conv2(out1)
out2 = pool(out2)
out2.shape

```

#Exercise 6: MNIST Tensor Calculator

```python
input = torch.Tensor(8, 64, 20, 23)
conv = nn.Conv2d(64, 128, kernel_size=(6, 5), stride=2, padding=1)
pool = nn.MaxPool2d(kernel_size=2, stride=3)
out1 = conv(input)
out1.size()
out2 = pool(out1)
out2.size()
```

```python
input = torch.Tensor(100, 64, 7, 7) #batch_size: 100, input_channel: 3, input_size: 28x28

conv = torch.nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1)

pool = torch.nn.AvgPool2d(kernel_size=1, stride=2)


out = conv(input)
out = pool(out)
out.size()

#input -> conv -> pool -> out
```

#Final Exam tensor calculator

```python
#layer 1

input = torch.Tensor(100, 1, 1000, 1000) #batch_size: 100, input_channel: 1, input_size: 28x28

conv = torch.nn.Conv1d(in_channels= 1, out_channels= 1, kernel_size= 2, stride= 2, padding= 0)

pool = torch.nn.MaxPool1d(kernel_size=1, stride=2)


out = conv(input)
out = pool(out)
out.size()

#input -> conv -> pool -> out
```
