# Question #13 Create categories by age as follows:

1. Between 0 and 18: Children
2. Between 19 and 25: Youth
3. Between 26 and 35: YoungAdult
4. Between 36 and 60: MiddleAged
5. Over 60: Senior

Then, find how many passengers belong to the Youth category

```python
Children = len(df.loc[(df.Age >= 0) & (df.Age < 19)])
Youth = len(df.loc[(df.Age >= 19) & (df.Age <= 25)])
YoungAdult = len(df.loc[(df.Age >= 26) & (df.Age <= 35)])
MiddleAged = len(df.loc[(df.Age >= 36) & (df.Age <= 60)])
Senior = len(df.loc[df.Age > 60])

age_range = pd.Series([Children, Youth, YoungAdult, MiddleAged, Senior],
                      index=['Children', 'Youth', 'YoungAdult', 'MiddleAged', 'Senior'])
age_range
```
