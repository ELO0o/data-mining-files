# 16. After extracting the data where the value of the item_name column is either 'steak salad' or 'bowl', remove duplicate rows based on that column.

> Hint: relational operator, logical operator, loc, and drop_duplicates(): https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop_duplicates.html

```python
df.loc[(df.item_name == 'Steak Salad') | (df.item_name == 'Bowl')].drop_duplicates('item_name')
```

# 17. Following Question 16 above, display only the last results under the same conditions

```python
df.loc[(df.item_name == 'Steak Salad') | (df.item_name == 'Bowl')].drop_duplicates('item_name', keep='last')
```
